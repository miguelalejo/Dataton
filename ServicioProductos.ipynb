```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cvs_compras/general.csv')
#df=pd.read_excel('cvs_compras/general.xlsx', index_col=0)
df.info()
df.head()
```

```python
df['RUC_EMPRESA'] = df['buyer.id'].str.slice(7, 20)
df['RUC_EMPRESA']
df['RUC_VENDEDOR'] = df['supplier_parties_identifier.id']
df['RUC_VENDEDOR']
```

```python
df_productos=pd.read_csv('cvs_compras/general_desglose_productos.csv')
df_productos.info()
df_productos.head()
```

```python
df.columns
```

```python
pd.set_option('display.max_columns', None)
df=df[['id', 'buyer.name',
       'buyer.id', 'tender.id', 
       'tender.resolution.description',
       'tender.resolution.period.startDate',
       'tender.resolution.period.endDate', 'awards_date',
       'awards_value.amount', 'contracts_id',           
       'contracts_period.durationInDays', 'contracts_value.amount',       
       'buyer_parties_address.streetAddress', 'buyer_parties_contactPoint.name',
       'buyer_parties_contactPoint.email',        
       'supplier_parties_identifier.legalName',
       'supplier_parties_address.streetAddress',              
       'supplier_parties_contactPoint.email',
        'RUC_EMPRESA', 'RUC_VENDEDOR']].copy()
```

```python
df_productos.columns
```

```python
df_productos = df_productos[['contracts_items_description',
       'contracts_items_quantity',
        'contracts_items_unit.name',
       'contracts_id']].copy()
```

```python
df_final = df.join(df_productos.set_index('contracts_id'), on='contracts_id')
```

```python
df_final.info()
df_final
```

```python
cantidad_productos_empresa = df_final.groupby(['RUC_EMPRESA','buyer.name','RUC_VENDEDOR','supplier_parties_identifier.legalName','contracts_items_description','contracts_items_unit.name']).agg({'contracts_items_quantity': ['sum']})
cantidad_productos_empresa.columns = ['total_productos']
cantidad_productos_empresa = cantidad_productos_empresa.reset_index()
df_cantidad_productos_empresa = cantidad_productos_empresa.copy()
display(df_cantidad_productos_empresa)
```

```python
def buscar_vendedor_por_ruc(ruc):
    return df_cantidad_productos_empresa.loc[df_cantidad_productos_empresa['RUC_VENDEDOR'] == ruc]
def buscar_empresa_por_ruc(ruc):
    return df_cantidad_productos_empresa.loc[df_cantidad_productos_empresa['RUC_EMPRESA'] == ruc]
```

```python
def buscar_productos_por_ruc(ruc):
    df_ruc = buscar_vendedor_por_ruc(ruc)
    df_top_productos = df_ruc.sort_values('total_productos',ascending = False).groupby('RUC_EMPRESA').head(8).copy()    
    return df_top_productos.head(8).copy()
```

```python
buscar_productos_por_ruc("1891734715001")
```

```python
from flask import Flask, request, render_template, session, redirect
from flask_cors import CORS
import json
app = Flask(__name__)
CORS(app, resources={r"/*": {"origins": "*"}})

@app.route('/', methods=("POST", "GET"))
def html_table():
    return render_template('simple.html',  tables=[df.to_html(classes='data')], titles=df.columns.values)

@app.route('/productos/rucs/<ruc>',methods=['GET'])
def buscar_ruc(ruc):         
    df_top_productos = buscar_productos_por_ruc(ruc)
    result = df_top_productos.to_json(orient="records")
    parsed = json.loads(result)    
    response = app.response_class(
        response=json.dumps(parsed, indent=4),
        status=200,
        mimetype='application/json'
    )
    return response

if __name__ == '__main__':
    app.run(host='0.0.0.0',port=5200)
```
