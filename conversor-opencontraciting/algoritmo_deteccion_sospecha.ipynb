### Lectura de los datos y procesamiento

```python
df=pd.read_csv('general.csv')
df['RUC_EMPRESA'] = df['buyer.id'].str.slice(7, 20)
df['RUC_EMPRESA']
df['RUC_VENDEDOR'] = df['supplier_parties_identifier.id']
df['RUC_VENDEDOR']
```

### Nombre de Columnas

```python
pd.set_option('display.max_columns', None)
df=df[['id', 'buyer.name',
       'buyer.id','tender.id', 
       'tender.resolution.description',
       'tender.resolution.period.startDate',
       'tender.resolution.period.endDate', 
       'awards_date',
       'awards_value.amount', 'contracts_id',           
       'contracts_period.durationInDays', 'contracts_value.amount',       
       'buyer_parties_address.streetAddress', 'buyer_parties_contactPoint.name',       
       'supplier_parties_identifier.legalName',        
        'RUC_EMPRESA', 'RUC_VENDEDOR']].copy()
df['fecha'] = df['awards_date'].astype('datetime64[ns]')
df.head(5)                 
```

### Buscamos los contratos que han durado 3 días o menos

```python
df['sospecha_dias'] = np.where(df['contracts_period.durationInDays']<3,"S","N")
```

### Visualizamos el RUc de los vendedores con más contratos

```python
df.loc[1]
rucs = df['RUC_VENDEDOR'].value_counts()
print(type(rucs))
print(type(rucs.to_frame()))

df_val = rucs.to_frame()
df_val.reset_index(inplace=True)
df_val.rename(columns={'RUC_VENDEDOR':'Contratos','index': 'RUC_VENDEDOR'}, inplace= True)
df_val.head(15)
```

### Detecamos los cambios bruscos en los montos, por ejemplo: si en la historia una empresa realiza contratos de aproximadamente de 10 000 USD, es algo sosprechoso que luego tenga contratos x20 de su valor, en este caso 200 000 USD.

```python
lista_step = []
def cambio_brusco(df_rucs):
    for i in range(len(df_rucs)):
        ruc = df_rucs.loc[i]
        df_contratos = df.loc[df['RUC_VENDEDOR'] == ruc["RUC_VENDEDOR"]]
        # display(df_contratos)
        lista_monto = []        
        contrato = df_contratos['awards_value.amount']
        lista_monto = list(contrato)
        # print(np.mean(lista_monto), np.median(lista_monto))
        Q3 = np.percentile(lista_monto, 75, interpolation = 'midpoint') 
        # print(Q3)
        # ruc['Q3'] = Q3  
        lista_step.append(Q3)

    df_rucs['sospecha_cambio_brusco'] = pd.Series(lista_step)         
    return df_rucs
    # df_val['sospecha_cambio_brusco'] = pd.Series(lista_step)
```

```python
df_1 = cambio_brusco(df_val)
```

```python
df_nuevo = pd.merge(df, df_1[['RUC_VENDEDOR', 'Contratos','sospecha_cambio_brusco']], how='inner', on='RUC_VENDEDOR')
```

```python
df_nuevo['sospecha_cambio'] = np.where(df_nuevo['awards_value.amount'] > df_nuevo['sospecha_cambio_brusco']*20, 'S', 'N')
```

### Si cumple los dos filtros lo marcamos como sospechoso

```python
datos_sospechosos = df_nuevo[(df_nuevo['sospecha_cambio'] == 'S') & (df_nuevo['sospecha_dias'] == 'S')].sort_values(by='contracts_value.amount', ascending=False)
```
