```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cvs_compras/general.csv')
df['RUC_EMPRESA'] = df['buyer.id'].str.slice(7, 20)
df['awards_date'] = df['awards_date'].str.slice(0, 10)

df['RUC_EMPRESA']
df['RUC_VENDEDOR'] = df['supplier_parties_identifier.id']
df['RUC_VENDEDOR']
df.info()
df.head()
```

```python
df.columns
```

```python
pd.set_option('display.max_columns', None)
df=df[['id', 'buyer.name',
       'buyer.id', 'tender.id', 
       'tender.resolution.description',
       'tender.resolution.period.startDate',
       'tender.resolution.period.endDate', 'awards_date',
       'awards_value.amount', 'contracts_id',           
       'contracts_period.durationInDays', 'contracts_value.amount',       
       'buyer_parties_address.streetAddress', 'buyer_parties_contactPoint.name',
       'buyer_parties_contactPoint.email',        
       'supplier_parties_identifier.legalName',
       'supplier_parties_address.streetAddress',              
       'supplier_parties_contactPoint.email',
        'RUC_EMPRESA', 'RUC_VENDEDOR']].copy()
df['fecha'] = df['awards_date'].astype('datetime64[ns]')
```

```python
ventas_empresa = df.groupby(['RUC_EMPRESA','buyer.name','RUC_VENDEDOR','supplier_parties_identifier.legalName','awards_date','fecha']).agg({'awards_value.amount': ['sum']})
ventas_empresa.columns = ['total_venta']
ventas_empresa = ventas_empresa.reset_index()
df_ventas_empresa = ventas_empresa.copy()
display(df_ventas_empresa)
```

```python
def buscar_vendedor_por_ruc(ruc):
    return df_ventas_empresa.loc[df_ventas_empresa['RUC_VENDEDOR'] == ruc]
def buscar_empresa_por_ruc(ruc):
    return df_ventas_empresa.loc[df_ventas_empresa['RUC_EMPRESA'] == ruc]
```

```python
def buscar_venta_fecha_por_ruc(ruc):
    df_ruc = buscar_vendedor_por_ruc(ruc)
    df_venta_fecha =  df_ruc.sort_values(['fecha','total_venta'],ascending = True).head(8).copy()
    return df_venta_fecha     
```

```python
from flask import Flask, request, render_template, session, redirect
from flask_cors import CORS
import json
app = Flask(__name__)
CORS(app, resources={r"/*": {"origins": "*"}})

@app.route('/', methods=("POST", "GET"))
def html_table():
    return render_template('simple.html',  tables=[df.to_html(classes='data')], titles=df.columns.values)

@app.route('/ventas-fechas/rucs/<ruc>',methods=['GET'])
def buscar_ruc(ruc):     
    
    df_top_productos = buscar_venta_fecha_por_ruc(ruc)
    result = df_top_productos.to_json(orient="records")
    parsed = json.loads(result)    
    response = app.response_class(
        response=json.dumps(parsed, indent=4),
        status=200,
        mimetype='application/json'
    )
    return response

if __name__ == '__main__':
    app.run(host='0.0.0.0',port=5800)
```
