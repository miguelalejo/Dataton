```python
import numpy as np
import pandas as pd
import codecs
```

```python
doc = codecs.open('rucs/GUAYAS.txt','rU','latin-1') #open for reading with "universal" type set
df = pd.read_csv(doc, sep='\t')
df = df.astype({'NUMERO_RUC': str}) 
df.info()
df.head()
```

```python
def buscar_por_ruc(ruc):
    return df.loc[df['NUMERO_RUC'] == ruc]
```

```python
buscar_por_ruc("0911729457001").info()
```

```python
from flask import Flask, request, render_template, session, redirect
from flask_cors import CORS
import json
app = Flask(__name__)
CORS(app, resources={r"/*": {"origins": "*"}})

@app.route('/', methods=("POST", "GET"))
def html_table():
    return render_template('simple.html',  tables=[df.to_html(classes='data')], titles=df.columns.values)

@app.route('/rucs/<ruc>',methods=['GET'])
def buscar_ruc(ruc): 
    df_ruc = buscar_por_ruc(ruc)
    result = df_ruc.to_json(orient="records")
    parsed = json.loads(result)    
    response = app.response_class(
        response=json.dumps(parsed, indent=4),
        status=200,
        mimetype='application/json'
    )
    return response

if __name__ == '__main__':
    app.run(host='0.0.0.0')
```
